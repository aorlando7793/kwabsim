```python
import os.path
import numpy as np
import math
import matplotlib.pyplot as plt
from cavityclass import *

LAM = 1064*10**(-7) * 1

#save_path = 'C:/Users/R&D Group/Python/kwabsim-master/'
#file_name = 'datatest'
#file_path = os.path.join(save_path, file_name+".dat")
```

```python
ds18 = Cavity('ds18.dat', LAM)
print(ds18.cavity)
xcav = ds18.get_xcav()
ycav = ds18.get_ycav()
```

```python
Zx, Wx = xcav.plot_waist(250)
Zy, Wy = ycav.plot_waist(250)
fig = plt.figure(figsize=(16,9))
ax = fig.add_subplot(1,1,1)
ax.plot(Zx,Wx,label='xcav')
ax.plot(Zy,Wy,label='ycav')
ax.set_xlabel('Z(cm)')
ax.set_ylabel('Spot Size(cm)')
ax.set_ylim(0,.3)
ax.legend()
plt.show()
```

```python
Wx_at_THG = []
Wx_at_V = []
Wy_at_THG = []
Wy_at_V = []
Z = []
for i in np.arange(10,30,.3):
    newcav = ds18.remove_optic(8).insert_optic('D',i,8)
    xcav2 = newcav.get_xcav()
    ycav2 = newcav.get_ycav()
    if xcav2.is_stable and ycav2.is_stable:
        Wx_at_THG.append(xcav2.waist(xcav2.L - 4.9))
        Wx_at_V.append(xcav2.waist(24.5))
        Wy_at_THG.append(ycav2.waist(ycav2.L - 4.9))
        Wy_at_V.append(ycav2.waist(24.5))
        Z.append(i)                   
fig2 = plt.figure(figsize=(16,9))
ax21 = fig2.add_subplot(2,1,1)
ax21.plot(Z, Wx_at_THG, 'r')
ax21.plot(Z, Wy_at_THG, 'g')
ax21.set_title('Spot Size at THG')

ax22 = fig2.add_subplot(2,1,2)
ax22.plot(Z, Wx_at_V, 'b')
ax22.plot(Z, Wy_at_V, 'k')
ax22.set_title('Spot Size at Nd:YVO_4')

plt.show()
```

```python
wVx = min(Wx_at_V)
index_Vx = Wx_at_V.index(wVx)
wTx = min(Wx_at_THG)
index_Tx = Wx_at_THG.index(wTx)

print('|Distance|','|Vanadate|', '|Thg|')
print(Z[index_Vx], wVx, Wx_at_THG[index_Vx])
print(Z[index_Tx],  Wx_at_V[index_Tx], wTx)
```
